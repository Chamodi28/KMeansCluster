```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
#print(iris)
X = iris.data
y = iris.target
```

```python

# Convert the Iris data to a Pandas DataFrame for easier correlation calculation
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Calculate the correlation matrix
corr_matrix = iris_df.corr()

# Generate the heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
iris_df.head()
```

```python
k_meansclus = range(1,10)
sse = []

for k in k_meansclus :
  km = KMeans(n_clusters =k)
  km.fit(X)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('Elbow Method')
plt.plot(k_meansclus,sse)
plt.show()
```

```python
KMeansCluster = KMeans(n_clusters=3,max_iter=300 , random_state=0)
KMeansCluster.fit(X)
```

```python
Y_means = KMeansCluster.fit_predict(X)
print(Y_means)
```

```python
df_imp = np.array(X)
df_imp
```

```python
plt.scatter(df_imp[Y_means==1,2 ],df_imp[Y_means==1,3 ], color='r' , label='Iris-setosa')
plt.scatter(df_imp[Y_means==2,2 ],df_imp[Y_means==2,3 ], color='b', label='Iris-virginica')
plt.scatter(df_imp[Y_means==0,2 ],df_imp[Y_means==0,3 ], color='g' , label='Iris-versicolor ')
plt.xlabel('petal length (cm)')
plt.ylabel('petal length (cm)')
plt.legend()
plt.title('Iris Species Scatter Plot')
plt.show()
```
